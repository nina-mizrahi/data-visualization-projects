```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flights=pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/nycflights13/flights.csv',index_col=0)
flights.head()
```

```python
fig,ax=plt.subplots()
ax=sns.histplot(data=flights,x='hour',binwidth=1,discrete=True,shrink=0.8)
ax.set_title('Number of flights from New York by hour')
ax.set_ylabel('Number')
fig.set_size_inches(7,5)
```

```python
sns.set_theme()
fig,ax=plt.subplots()
temp = data=flights[(flights.month==7)|(flights.month==8)|(flights.month==9)]
temp['day']=temp['time_hour'].apply(lambda x:pd.to_datetime(x).strftime('%A'))
sns.histplot(data=temp,x='day',hue='month',multiple='dodge',palette='muted',shrink=0.8)
ax.legend(labels=['July','August','September'])
ax.set_title('Number of flights on each weekday in July, August, and September')
fig.set_size_inches(10,8)
```

```python
sns.set_style("dark")
fig,ax=plt.subplots()
ax=sns.histplot(data=flights,x='carrier',binwidth=1,discrete=True,shrink=1,hue='origin',multiple='stack',palette='dark')
ax.set_title('Number of flights from each airport, by carrier')
fig.set_size_inches(8,6)
```

```python
perceewr=flights[flights.origin=='EWR']
perceewr
```
