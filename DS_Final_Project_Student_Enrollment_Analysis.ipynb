```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/enrollmentdata.csv')
df
```

```python
def generateDisciplines(row):
  if row['Disc']=='SPAN' or row['Disc']=='KORE' or row['Disc']=='FREN' or row['Disc']=='GERM' or row['Disc']=='JAPN'or row['Disc']=='RUSS'or row['Disc']=='ITAL' or row['Disc']=='CHIN' or row['Disc']=='LING' or row['Disc']=='FLAN':
    return 'Language Studies'
  elif row['Disc']=='HIST':
    return 'History'
  elif row ['Disc']=='ENGR' or row['Disc']=='FYS':
    return 'Engineering'
  elif row['Disc']=='CSCI' or row['Disc']=='CSMT':
    return 'Computer Science'
  elif row['Disc']=='MATH' or row['Disc']=='STS':
    return 'Mathematics'
  elif row['Disc']=='PHIL' or row['Disc']=='LGSC' or row['Disc']=='HSID':
    return 'Philosophy'
  elif row['Disc']=='BIOL' or row['Disc']=='CHEM' or row['Disc']=='PHYS' or row['Disc']=='ASTR' or row['Disc']=='GEOL' or row['Disc']=='NEUR' or row['Disc']=='MOBI':
    return 'Science'
  elif row['Disc']=='PSYC':
    return 'Psycology'
  elif row['Disc']=='ECON' or row['Disc']=='ORST':
    return 'Economics'
  elif row['Disc']=='GOVT' or row['Disc']=='POLI' or row['Disc']=='POST' or row['Disc']=='PPE' or row['Disc']=='LGST' or row['Disc']=='IR':
    return 'Government'
  elif row['Disc']=='LIT' or row['Disc']=='ENGL' or row['Disc']=='WRIT':
    return 'Literature'
  elif row['Disc']=='RLST' or row['Disc']=='JWST':
    return 'Religious Studies'
  elif row['Disc']=='ART' or row['Disc']=='ARHI':
    return 'Art and Art History'
  elif row['Disc']=='ANTH' or row['Disc']=='IIS' or row['Disc']=='SOC':
    return 'Anthropology'
  elif row['Disc']=='DANC' or row['Disc']=='MS' or row['Disc']=='THEA' or row['Disc']=='MUS' or row['Disc']=='CREA':
    return 'Creative Studies'
  elif row['Disc']=='ENVS' or row['Disc']=='EA':
    return 'Enviromental Studies'
  elif row['Disc']=='GFS' or row['Disc']=='WMST' or row['Disc']=='GWS' or row['Disc']=='SPSY' or row['Disc']=='WGFS':
    return 'Gender Studies'
  else:
    return 'Other'
```

```python
data=df
data['Discipline']=data.apply(generateDisciplines, axis=1)
data
```

```python
data[data['Discipline']=='Other'].iloc[10:40,0:12]
```

```python
data=(data[data['AcadYr']=='2020-21']. #filter out 2020-21 academic year
      rename(columns={' CM':'CM',' HM':'HM',' PO':'PO',' PZ':'PZ',' SC':'SC'}). #rename columns
      groupby(by='Discipline').sum()) #groupby and sum
data
```

```python
CMdata=data.CM.sort_values(ascending=False)
CMdata
```
