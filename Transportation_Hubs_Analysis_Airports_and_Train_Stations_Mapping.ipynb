```python
import plotly.express as px
import pandas as pd
```

```python
airports=pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/nycflights13/airports.csv')
airports
```

```python
airalts=airports[airports.name.str.contains('Station')==False]
airalts
```

```python
px.scatter_mapbox(airports,
                  lat='lat',
                  lon='lon',
                  mapbox_style='open-street-map',
                  zoom=2.5,
                  color='alt',
                  color_continuous_scale='icefire',
                  opacity=0.5,
                  title='Airports Altitudes',
                  hover_data={'lat':False,'lon':False,'alt':False,'name':True})
```

```python
trainzones=airports[airports.name.str.contains('Station')]
trainzones.head()
```

```python
trainzones=trainzones[trainzones.tzone.str.contains('New_York')]
trainzones
```

```python
px.scatter_mapbox(trainzones,
                  lat='lat',
                  lon='lon',
                  mapbox_style='open-street-map',
                  zoom=5,
                  title='Train stations in the NY time zone',
                  hover_data={'lat':False,'lon':False,'alt':False,'name':True})
```

```python
splitname=(airports[airports.tzone == 'America/New_York'].
           name.str.rsplit(None, 1, expand = True).
           rename(columns = {1:'type'}))
splitname
```

```python
airtrain = splitname.merge(airports, how = 'outer', left_index = True, right_index = True).dropna()
airtrain
```

```python
newairtrain=airtrain[(airtrain.type=='Airport')|(airtrain.type=='Station')]
```

```python
px.scatter_mapbox(newairtrain,
                  lat='lat',
                  lon='lon',
                  mapbox_style='carto-positron',
                  zoom=5,
                  color='type',
                  title='Airports and train stations in the NY Time Zone',
                  hover_data={'lat':False,'lon':False,'alt':False,'name':True})
```
