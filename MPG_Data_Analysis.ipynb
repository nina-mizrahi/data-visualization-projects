```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
mpg=pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/ggplot2/mpg.csv')
mpg.head()
```

```python
sns.set_theme(style='darkgrid')
fig,ax=plt.subplots()
ax=sns.scatterplot(data=mpg,x='displ',y='cty',style='cyl',hue='cyl',palette='deep')
ax.set_xlabel('Engine Displacement')
ax.set_ylabel('City MPG')
ax.set_title('City Mileage vs Displacement')
fig.set_size_inches((7,5))
```

```python
sns.set_theme(style='darkgrid')
fig,ax=plt.subplots()
ax=sns.scatterplot(data=mpg,x='cty',y='hwy',style='year',hue='class',palette='deep')
ax.set_xlabel('City MPG')
ax.set_ylabel('Highway MPG')
ax.set_title('City vs Highway mileage for various car classes')
fig.set_size_inches((8,6))
```

```python
sns.set_theme(style='darkgrid')
fig,ax=plt.subplots()
ax=sns.scatterplot(data=mpg,x='class',y='displ',size='cty',sizes=(10,300))
ax.set_xlabel('Class')
ax.set_ylabel('Displacement')
ax.set_title('How displacement effects mileage')
fig.set_size_inches((8,6))
```
