```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
netflix = pd.read_csv('https://drive.google.com/uc?id=11h0mCiHAuBT3GVoJSOIIYC0nov62x7vP',index_col=0)
netflix.head(5)
```

```python
sns.set_theme()
sns.histplot(data=netflix,x='type',shrink=0.8,palette='colorblind')
```

```python
sns.histplot(data=netflix[netflix.release_year>=2000],x='release_year',shrink=1,palette='colorblind',hue='type',multiple='dodge')
```

```python
sns.histplot(data=netflix[netflix.release_year>=2000],x='release_year',shrink=1,palette='colorblind',hue='type',multiple='stack')
```

```python

sns.histplot(data=netflix[netflix.release_year>=2000],x='release_year',shrink=1,palette='colorblind',multiple='stack')
```

```python
fig,ax=plt.subplots()
ax=sns.histplot(data=netflix[netflix.release_year>=2000],
             x='release_year',
             shrink=1,
             palette='colorblind',
             hue='type',
             multiple='stack')
ax.set_title('Netflix data by release year')
ax.legend(labels=['TV','Movie'])
fig.set_size_inches(8,6)
```

```python
sns.histplot(data=netflix[netflix.release_year>=2000],
             x='release_year',
             stat='probability')
```

```python
movies=netflix[netflix.type=="Movie"]
```

```python
movies
```

```python
movies['time']=movies.duration.str.findall(r'\d+').apply(lambda x:x[0]).astype(float)
```

```python
movies
```

```python
sns.histplot(data=movies[movies.time<200],x='time',binwidth=10)
```

```python
sns.histplot(data=movies[movies.time<200],x='time',binwidth=10,stat='percent')
```

```python
sns.histplot(data=netflix,x='rating')
```

```python
counts=netflix.groupby(by='rating').size().sort_values(ascending=False)
counts
```

```python
explodelist=[0]*len(counts)
explodelist[0]=0.1
```

```python
palette=sns.color_palette('pastel')
fig,ax=plt.subplots()
ax=plt.pie(x=counts,autopct='%1.1f%%',colors=palette,explode=explodelist,shadow=True)
plt.legend(labels=counts.index)
fig.set_size_inches(10,10)
```
