```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
weather=pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/nycflights13/weather.csv')
weather.head()
```

```python
noontemp=weather[weather.hour==12][weather.origin=='EWR']
noontemp
```

```python
sns.set_theme()
fig,ax=plt.subplots()
ax=sns.lineplot(data=noontemp,x='month',y='temp')
ax.set_title('Temperature at noon at EWR by month')
fig.set_size_inches(10,8)
```

```python
weekweather=weather[weather.origin=='EWR'].assign(week=lambda x:x.month*4)
weekweather
```

```python
sns.set_theme()
fig,ax=plt.subplots()
ax=sns.lineplot(data=weekweather,x='week',y='temp',ls='--')
ax.set_title('Temperature at EWR by week')
fig.set_size_inches(10,8)
```

```python
hour=[0,12]
augusttemp=weather[weather.month==8][weather['hour'].isin(hour)]
augusttemp
```

```python
sns.set_theme()
fig,ax=plt.subplots()
ax=sns.lineplot(data=augusttemp,x='day',y='temp',size='hour',ci=None)
ax.set_title('August Temperatures at noon and midnight')
fig.set_size_inches(10,8)
```

```python
firstnoon=weather[weather.hour==12][weather.day==1]
firstnoon
```

```python
weather['time_hour'] = pd.to_datetime(weather['time_hour']).apply(lambda x:x.strftime('%y-%d'))
```

```python
#weather['date']=weather.time_hour.dropna().apply(lambda x:x.strftime('%y-%d')) was giving me an error
sns.set_theme()
fig,ax=plt.subplots()
ax=sns.lineplot(data=firstnoon,x='time_hour',y='temp',ci=None,hue='origin',style='origin',markers=True,markersize=8)
ax.set_title('Temperatures at noon on the first day of each month')
fig.set_size_inches(8,6)
```

```python
firstday=weather[weather.day==1][weather.hour==12]
```

```python
sns.set_theme()
fig,axes=plt.subplots(2,sharex=True)
sns.lineplot(data=firstday,x='month',y='temp',ci=None,ax=axes[0])
sns.lineplot(data=firstday,x='month',y='humid',ci=None,ax=axes[1])
axes[0].set_title('First day of the month at noon')
fig.set_size_inches(9,9)
```
