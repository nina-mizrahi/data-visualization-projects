```python
import pandas as pd
```

```python
import plotly.express as px
```

```python
df = px.data.carshare()
```

```python
df
```

```python
px.scatter_mapbox(df,lat='centroid_lat',lon='centroid_lon',mapbox_style='carto-positron')
```

```python
px.scatter_mapbox(df,lat='centroid_lat',lon='centroid_lon',mapbox_style='carto-positron',zoom=10)
```

```python
px.scatter_mapbox(df,
                  lat='centroid_lat',
                  lon='centroid_lon',
                  mapbox_style='carto-positron',
                  zoom=10,
                  color='car_hours',
                  color_continuous_scale='icefire')
```

```python
px.scatter_mapbox(df,
                  lat='centroid_lat',
                  lon='centroid_lon',
                  mapbox_style='open-street-map',
                  zoom=12,
                  color='car_hours',
                  color_continuous_scale='icefire',
                  size='peak_hour',
                  opacity=0.5,
                  hover_data={'centroid_lat':False,'centroid_lon':False},
                  title='Car Data')
```
